```python
df_notebooks_noTagged=pd.read_pickle('ASHRAE-Kaggle_notebooks_meta.pkl')
df_notebooks=pd.read_excel('ASHRAE-Kaggle_notebooks_meta(Tagged).xlsx')
df_notebooks = df_notebooks.drop(['title','tags','data'],axis=1).merge(df_notebooks_noTagged[['title','scriptUrl', 'tags', 'data']], on='scriptUrl')
df_notebooks
```

```python
colName='languageName'
title='Script language'

df_temp = df_notebooks[colName].value_counts().sort_values(ascending=False)
df_temp = pd.DataFrame(df_temp)
df_temp.columns = ['Count']
df_temp[colName] = df_temp.index
df_temp.iplot(kind='pie', labels=colName, values='Count', title=title, textinfo="label+percent", hole=0.3, theme='white')
```

```python
colName='isGpuEnabled'
title='Is GPU enabled?'

df_temp = df_notebooks[colName].value_counts().sort_values(ascending=False)
df_temp = pd.DataFrame(df_temp)
df_temp.columns = ['Count']
df_temp[colName] = df_temp.index
df_temp.iplot(kind='pie', labels=colName, values='Count', title=title, textinfo="label+percent", hole=0.3, theme='white')
```

```python
colName='Models\t'
title='Which model is used?'

df_temp = df_notebooks[colName].value_counts().sort_values(ascending=False)
df_temp = pd.DataFrame(df_temp)
df_temp.columns = ['Count']
df_temp[colName] = df_temp.index
df_temp.iplot(kind='pie', labels=colName, values='Count', title=title, textinfo="label+percent", hole=0.3, theme='white')
```

```python
list_tags = np.reshape(df_notebooks['tags'].to_list(), (1,-1)).tolist()[0]
dict_tags = {tag:list_tags.count(tag) for tag in list_tags}
df_tags_count = pd.DataFrame.from_dict(dict_tags, orient='index').sort_values(0, ascending=False).drop('nan')
df_tags_count = df_tags_count.reset_index().rename(columns={'index':'tag', 0:'Count'})
df_tags_count = df_tags_count[df_tags_count['Count']>1]
df_tags_count
```

```python
title = 'Tags of notebooks'
df_tags_count.iplot(kind='pie', labels='tag', values='Count', title=title, textinfo="label+percent", hole=0.3, theme='white')
```

```python
df_tags_count.set_index('tag').iplot(kind='bar', yTitle='Number of tags', title='Tags of notebooks', theme='white',
                    filename='cufflinks/categorical-bar-chart')
```

```python
df_top50 = df_notebooks[['title','totalVotes', 'bestPublicScore', 'Leak_data_usage']].sort_values('totalVotes', ascending=False).head(50)
df_top50.loc[df_top50['bestPublicScore']>0,'category'] = 'Prediction Models'
df_top50.loc[(df_top50['category'].isna()) & (df_top50['Leak_data_usage']==1),'category'] = 'Leak Data'
df_top50.loc[(df_top50['category'].isna()) & (df_top50['title'].str.contains('EDA|Explor|Introduction')),'category'] = 'EDA'
df_top50.loc[(df_top50['category'].isna()),'category'] = 'Preprocessing'
df_top50
```

```python
df_notebooks.sort_values('totalVotes', ascending=False)[['title','totalVotes','medal']].reset_index(drop=True)\
            .replace('gold','ðŸ¥‰(gold)').replace('silver','ðŸ¥ˆ(silver)').replace('bronze','ðŸ¥‡(bronze)').head(50)
```

```python
title='Categories of Top50'

df_concat = pd.DataFrame()
for category in ['EDA', 'Leak Data', 'Prediction Models', 'Preprocessing']:
    df_concat = pd.concat([df_concat, pd.DataFrame(df_top50[df_top50['category']==category]['totalVotes'].reset_index(drop=True).rename(category))], axis=1)
df_concat.iplot(kind='box',title=title,yTitle='Total votes', theme='white')
```

```python
title='Categories of Top50'
df_top50.groupby('category').count()['title'].reset_index().iplot(kind='pie', labels='category', values='title', title=title, textinfo="label+percent", hole=0.3, theme='white')
```
