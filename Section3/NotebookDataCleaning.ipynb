```python
import pandas as pd
import ast
import json
import requests
```

# Scrape notebooks data (keyword: 'ASHRAE')

```python
#Get the json data via webAPI
df_notebooks = pd.read_json('https://www.kaggle.com/kernels.json?search=ASHRAE&pageSize=1000')

#Handle datetime columns
df_notebooks['scriptVersionDateCreated'] = pd.to_datetime(df_notebooks['scriptVersionDateCreated'])

df_notebooks
```

# Clean notebooks data

```python
#Get column of tags
idx_item = 0
while True:
    try:
        df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[idx_item].apply(pd.Series)['name']], axis=1)
        idx_item = idx_item+1
    except:
        break
df_notebooks['tags'] = pd.Series((df_notebooks['name'].values.tolist()))
df_notebooks = df_notebooks.drop('name', axis=1)

idx_item = 0
while True:
    try:
        df_notebooks = pd.concat([df_notebooks, df_notebooks['dataSources'].apply(pd.Series)[idx_item].apply(pd.Series)['name']], axis=1)
        idx_item = idx_item+1
    except:
        break
df_notebooks['data'] = pd.Series((df_notebooks['name'].values.tolist()))
df_notebooks = df_notebooks.drop('name', axis=1)

#Filtering: only notebooks
df_notebooks = df_notebooks[df_notebooks['isNotebook']==True]

#Leave useful columns(meta)
df_notebooks = df_notebooks[['title','scriptUrl','tags','data','languageName','bestPublicScore'
                             ,'isGpuEnabled','lastRunExecutionTimeSeconds','medal','totalComments', 
                             'totalForks', 'totalViews', 'totalVotes']]
df_notebooks
```

```python
#combline for complete Url
df_notebooks['scriptUrl'] = 'https://www.kaggle.com'+df_notebooks['scriptUrl']

'''
#One-hot encoding for tags
df_notebooks = df_notebooks.join(df_notebooks['tags'].str.join('|').str.get_dummies())
df_notebooks = df_notebooks.join(df_notebooks['data'].str.join('|').str.get_dummies())
'''

df_notebooks
```

```python
#Remove non-ASHRAE notebooks
df_notebooks = df_notebooks[((df_notebooks['data'].astype('str').str.contains('ASHRAE', case=False)) | (df_notebooks['title'].str.contains('ASHRAE', case=False)))]
df_notebooks
```
