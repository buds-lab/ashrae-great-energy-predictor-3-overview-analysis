```python
import pandas as pd
import ast
import json
import requests
```

# Scrape notebooks data (keyword: 'ASHRAE')

```python
#Get the json data via webAPI
df_notebooks = pd.read_json('https://www.kaggle.com/kernels.json?search=ASHRAE&pageSize=1000')

#Handle datetime columns
df_notebooks['scriptVersionDateCreated'] = pd.to_datetime(df_notebooks['scriptVersionDateCreated'])

df_notebooks
```

# Clean notebooks data

```python
#Get column of tags
df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[0].apply(pd.Series)['name']], axis=1)
df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[1].apply(pd.Series)['name']], axis=1)
df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[2].apply(pd.Series)['name']], axis=1)
df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[3].apply(pd.Series)['name']], axis=1)
df_notebooks = pd.concat([df_notebooks, df_notebooks['categories'].apply(pd.Series)[4].apply(pd.Series)['name']], axis=1)
df_notebooks['tags'] = pd.Series((df_notebooks['name'].values.tolist()))
df_notebooks = df_notebooks.drop(['categories', 'name'], axis=1)

#Filtering: only notebooks
df_notebooks = df_notebooks[df_notebooks['isNotebook']==True]

#Leave useful columns(meta)
df_notebooks = df_notebooks[['title','scriptUrl','languageName','bestPublicScore','isGpuEnabled',
                             'lastRunExecutionTimeSeconds','medal','totalComments', 
                             'totalForks', 'totalViews', 'totalVotes','tags']]
df_notebooks
```

```python
#combline for complete Url
df_notebooks['scriptUrl'] = 'https://www.kaggle.com'+df_notebooks['scriptUrl']

#One-hot encoding for tags
df_notebooks = df_notebooks.drop('tags', axis=1).join(df_notebooks['tags'].str.join('|').str.get_dummies())

df_notebooks
```

```python
df_notebooks.to_csv('ASHRAE-Kaggle_notebooks_meta.csv', index=False)
```
