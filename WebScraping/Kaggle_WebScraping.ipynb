```python
import pandas as pd

import json
import requests
```

# Scrape notebooks data (keyword: 'ASHRAE')

```python
#Get the json data via webAPI
df_notebooks = pd.read_json('https://www.kaggle.com/kernels.json?search=ASHRAE&pageSize=1000')

#Handle datetime columns
df_notebooks['scriptVersionDateCreated'] = pd.to_datetime(df_notebooks['scriptVersionDateCreated'])

df_notebooks
```

```python
df_notebooks.info()
```

# Scrape discussion data (keyword: 'ASHRAE')

```python
df_discussion = pd.DataFrame()
page = 0

#Scrape data page by page
while True:
    try:
        print('page: ' + str(page))
        
        #Get the json data via webAPI
        df_temp = pd.read_json('https://www.kaggle.com/topics.json?search=ASHRAE&page=' + str(page))
        
        df_discussion = df_discussion.append(df_temp, ignore_index=True)
        
        page = page+1
        
        print('Sucessful')
        print('----------')
        
    except:
        print('Fail')
        break

#Handle dict column
df_discussion = pd.concat([df_discussion.drop(['topics'], axis=1), df_discussion['topics'].apply(pd.Series)], axis=1)

#Handle datetime columns
df_discussion['lastCommentPostDate'] = pd.to_datetime(df_discussion['lastCommentPostDate'])
df_discussion['postDate'] = pd.to_datetime(df_discussion['postDate'])        
        
df_discussion
```

```python
df_discussion.info()
```

# Scrape competition data (only completed ones)

```python
df_competitions = pd.DataFrame()
page = 0

#Scrape data page by page
while True:
    try:
        print('page: ' + str(page))
        
        #Get the json data via webAPI
        data = json.loads(requests.get('https://www.kaggle.com/competitions.json?sortBy=grouped&group=general&page=' + str(page)).content)
        df_temp = pd.DataFrame(data['pagedCompetitionGroup']['competitions'])
        
        #Handle dict column
        df_temp = pd.concat([df_temp.drop(['categories'], axis=1), df_temp['categories'].apply(pd.Series)], axis=1)
        
        df_competitions = df_competitions.append(df_temp, ignore_index=True)
        
        page = page+1
        
        print('Sucessful')
        print('----------')
        
    except:
        print('Fail')
        break
        
#Handle datetime columns
df_competitions['deadline'] = pd.to_datetime(df_competitions['deadline'])
df_competitions['enabledDate'] = pd.to_datetime(df_competitions['enabledDate'])
df_competitions['mergerDeadline'] = pd.to_datetime(df_competitions['mergerDeadline'])
df_competitions['organizationAvatarUrl'] = pd.to_datetime(df_competitions['organizationAvatarUrl'])

df_competitions
```

```python
df_competitions.info()
```

```python
with pd.ExcelWriter('ASHRAE-Kaggle_WebScraping.xlsx',engine='xlsxwriter',options={'remove_timezone': True}) as writer:  # doctest: +SKIP
    df_notebooks.to_excel(writer, sheet_name='notebooks')
    df_discussion.to_excel(writer, sheet_name='discussion')
    df_competitions.to_excel(writer, sheet_name='competitions')    
```
